## Hypothesis 3: People with a higher personal income per week tend to live in areas near/along the coast.


```python
#2011 Data
import pandas as pd
Income_2011A = pd.read_csv("2011Census_B17A_WA_POA_short.csv") #index_col = 0)
```

```python
Income_2011B = pd.read_csv("2011Census_B17B_WA_POA_short.csv") #index_col = 0)
```

```python
Income_2011 = pd.merge(Income_2011A, Income_2011B, left_on="region_id", right_on="region_id")
```

```python
Income_2011.head()
```

```python
Income_2011.columns.get_loc('P_Neg_Nil_income_Tot')
```

```python
Income_2011.columns.get_loc('P_2000_more_Tot')
```

```python
Income_2011_6010 = Income_2011.iloc[8, 270:371:10]
Income_2011_6010
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20, 10))
Income_2011_6010.plot(kind='bar', color = "green")

plt.xlabel("Personal Income Per Week")
plt.ylabel("Frequency")
plt.title("2011 Distribution of Personal Incomes Per Week in Claremont")
plt.show()
```
