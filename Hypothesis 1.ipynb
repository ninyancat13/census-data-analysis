```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20, 10))
Age_2011_6000.plot(kind='bar')

plt.xlabel("Age")
plt.ylabel("Number of Residents")
plt.title("2011 Age Distribution of People living in Perth City")
plt.show()
```
